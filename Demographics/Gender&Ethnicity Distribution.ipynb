```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sim_av_patient.csv dataset
patient_df = pd.read_csv("C:/Users/wardo/Documents/Internship/cancer_download/Data/sim_av_patient.csv", low_memory=False)

# Map gender for better readability
gender_mapping = {1: 'Male', 2: 'Female', 9: 'Unknown'}
patient_df['GENDER_mapped'] = patient_df['GENDER'].map(gender_mapping)

# Simple filters for gender and ethnicity
# Uncomment and modify filters as needed
# Example: Filter for a specific gender (e.g., 'Male')
# patient_df = patient_df[patient_df['GENDER_mapped'] == 'Male']

# Example: Filter for a specific ethnicity (e.g., 'A')
# patient_df = patient_df[patient_df['ETHNICITY'] == 'A']

# Plot gender distribution
gender_counts = patient_df['GENDER_mapped'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'orange', 'grey'], alpha=0.7)
plt.title("Gender Distribution", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=0)
plt.show()

# Plot ethnicity distribution (top 10 most frequent ethnicities)
ethnicity_counts = patient_df['ETHNICITY'].value_counts().head(10)
plt.figure(figsize=(10, 6))
ethnicity_counts.plot(kind='bar', color='purple', alpha=0.7)
plt.title("Top 10 Ethnicity Distribution", fontsize=16)
plt.xlabel("Ethnicity", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
