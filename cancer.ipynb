```python
import pandas as pd

# Replace with the path to your dataset
file_path = "C:/Users/wardo/Documents/Internship/ProjectCancerData/sim_av_tumour.csv"
data = pd.read_csv(file_path)

# View the structure of the dataset
print(data.info())

# Preview the first few rows
print(data.head())
```

```python
# Summary statistics for numeric columns
print(data.describe())

# Summary statistics for non-numeric columns
print(data.describe(include='object'))
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numeric columns
data.hist(bins=20, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Bar plots for categorical columns
categorical_cols = data.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data[col])
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import numpy as np

# Exclude non-numeric columns before calculating the correlation
numeric_data = data.select_dtypes(include=[np.number])

# Calculate the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Export summary statistics to a CSV file
summary = data.describe()
summary.to_csv("summary_statistics.csv")
```
